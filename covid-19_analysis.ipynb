# Apple Mobility data set

### Find the change in driving mobility as time goes by (date increases)

```python
%%bigquery
select region, date, change_in_baseline_mobility
from nimble-cortex-266516.covid_19_apple_mobility_modeled.mobility_report_beam_DF
where transportation_type = 'driving' and region = 'United States'
order by date asc
```

```python
%%bigquery
create or replace view `nimble-cortex-266516.covid_19_views.v_driving_mobility` as
select region, date, change_in_baseline_mobility
from `nimble-cortex-266516.covid_19_apple_mobility_modeled.mobility_report_beam_DF`
where transportation_type = 'driving'
order by date asc
```

```python
%%bigquery
select date, avg(change_in_baseline_mobility) as average_change_among_countries
from nimble-cortex-266516.covid_19_apple_mobility_modeled.mobility_report_beam_DF
where transportation_type = 'driving' and geo_type = 'country/region'
group by date
order by date asc
```

```python
%%bigquery
create or replace view `nimble-cortex-266516.covid_19_views.v_driving_mobility_avg` as
select date, avg(change_in_baseline_mobility) as average_change_among_countries
from `nimble-cortex-266516.covid_19_apple_mobility_modeled.mobility_report_beam_DF`
where transportation_type = 'driving' and geo_type = 'country/region'
group by date
order by date asc
```

# Google Mobility and USAFacts county join

```python
%%bigquery
select g.sub_region_1 as state, g.sub_region_2 as county, g.date, g.retail_and_recreation_percent_change_from_baseline, g.grocery_and_pharmacy_percent_change_from_baseline, g.workplaces_percent_change_from_baseline, u.confirmed_cases, u.deaths
from nimble-cortex-266516.covid_19_google_mobility_modeled.mobility_report_Beam_DF g join nimble-cortex-266516.covid_19_usa_staging.us_counties u
on g.date = u.date and g.sub_region_1 = u.state_name and g.sub_region_2 = u.county
where g.country_region = 'United States' and g.sub_region_1 = 'Texas' and g.sub_region_2 = 'Travis'
order by g.date asc
```

```python
%%bigquery
create or replace view `nimble-cortex-266516.covid_19_views.v_usa_cases_and_mobility` as
select g.sub_region_1 as state, g.sub_region_2 as county, g.date, g.retail_and_recreation_percent_change_from_baseline, g.grocery_and_pharmacy_percent_change_from_baseline, g.workplaces_percent_change_from_baseline, u.confirmed_cases, u.deaths
from `nimble-cortex-266516.covid_19_google_mobility_modeled.mobility_report_Beam_DF` g join `nimble-cortex-266516.covid_19_usa_staging.us_counties` u
on g.date = u.date and g.sub_region_1 = u.state_name and g.sub_region_2 = u.county
where g.country_region = 'United States'
order by g.date asc
```

# JHU summary and world cities join

```python
%%bigquery
select s.date, w.city, s.province, s.country_region, s.confirmed, s.deaths, s.recovered, s.active, (s.confirmed/w.population*100) as confirmed_per_100, (s.deaths/w.population*100) as deaths_per_100, (s.recovered/w.population*100) as recovered_per_100, (s.active/w.population*100) as active_per_100
from nimble-cortex-266516.covid_19_jhu_modeled.summary_Beam_DF s join nimble-cortex-266516.world_cities_staging.world_cities w
on s.admin2 = w.city_ascii and s.province = w.admin_name and s.country_region = w.country
where s.country_region = 'United States'
order by s.date asc
```

```python
%%bigquery
create or replace view `nimble-cortex-266516.covid_19_views.v_usa_cases_to_population` as
select s.date, w.city, s.province, s.country_region, s.confirmed, s.deaths, s.recovered, s.active, (s.confirmed/w.population*100) as confirmed_per_100, (s.deaths/w.population*100) as deaths_per_100, (s.recovered/w.population*100) as recovered_per_100, (s.active/w.population*100) as active_per_100
from `nimble-cortex-266516.covid_19_jhu_modeled.summary_Beam_DF` s join `nimble-cortex-266516.world_cities_staging.world_cities` w
on s.admin2 = w.city_ascii and s.province = w.admin_name and s.country_region = w.country
where s.country_region = 'United States'
order by s.date asc
```
